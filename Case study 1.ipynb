### Importing necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

### 1. Reading the dataset

```python
mydata = pd.read_csv("cars_data.csv")
mydata
```

### 2. Checking for the null values in the dataset

```python
mydata.isnull()
```

### 3. Plotting bar graph of male vs female buyers participated in the sales

```python
mydata['Car Gender'].value_counts().plot(kind='bar', figsize=(7, 6), rot=0)
mydata['Buyer Gender'].value_counts().plot(kind='bar', figsize=(7, 6), rot=0)
plt.xlabel("Gender")
plt.ylabel("Count of People")
plt.title('Male vs Female buyers participated in the sales')
```

```python
sorted_salesprice = mydata.sort_values(['Sale Price'], ascending=False)
top = sorted_salesprice[['Make','Model','Sale Price']].head()
top
```
